## Pandas Refersher

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Importing the data from different file types
df_csv = pd.read_csv('pokemon.csv')
df_xls = pd.read_excel('pokemon_data.xlsx')
df_txt = pd.read_csv('pokemon_data.txt',delimiter = "\t")
print(df_csv.head())
print(df_xls.head())
print(df_txt.head())
```

## Reading Pandas data

```python
print(df_csv.columns)
print(df_csv['Name'][0::2])
print(df_csv[['Name','Legendary']])
#Use of iloc
print(df_csv.iloc[0::2])
print(df_csv.iloc[0::2]['Legendary'])
```

```python
#Use of loc is to locate based on condition
print(df_csv.loc[df_csv['Legendary']==True])
```

## Sorting/Describing Data

```python
print(df_csv.sort_values('Name',ascending = False))
print(df_csv.sort_values(['Name','HP'],ascending = [True,True]))
```

```python
#Print the statistical datas
df_csv.describe()
```

```python
# Print the count and datatupe for each column
df_csv.info()
```

```python
# Modifying data
# To drop columns
df_csv = df_csv.drop(columns=['Sp. Atk', 'Sp. Def'])
print(df_csv.head())
```

## Saving Pandas Data

```python
#Save to csv
df_csv.to_csv('pokemon_modified.csv',index=False)
#Save to excel
df_csv.to_excel('pokemon_modified.xlsx',index=False)
#Save to txt
df_csv.to_csv('pokemon_modified.txt',index=False,sep='\t')
```

## Filtering data

```python
#Based on multiple conditios
print(df_csv.loc[(df_csv['Type 1'] == 'Grass') & (df_csv['Legendary']==True)].reset_index())
```

```python
#Regex filtering
#print(df_csv)
print(df_csv.loc[df_csv['Name'].str.contains('Mega')])
print(df_csv.loc[~df_csv['Name'].str.contains('Mega')])
```

## Groupby Statistics

```python
print(df_csv.groupby(['Legendary']).count())
print(df_csv.groupby(['Type 1']).value_counts())
```
