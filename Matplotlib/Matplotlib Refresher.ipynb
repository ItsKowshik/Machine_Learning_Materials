```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Plot scatterplot
np.random.seed(42)
x = np.random.random(50)*100
y = np.random.random(50)*100
#plt.scatter(x,y)
#Same plot with different color
#plt.scatter(x,y,c='r')
#Same plot with different color and marker
#plt.scatter(x,y,c='r',marker='+')
#Same plot with different color, marker and size
plt.scatter(x,y,c='blue',marker='*',s=100)
```

```python
#Plot lineplot
years = [2006 +x for x in range(16)]
print(years)
weights = [50, 54, 56, 52, 58, 57 , 51 , 60, 62, 61, 66, 63, 59, 55, 65, 57]
#plt.plot(years,weights)
#Same plot with different color
#plt.plot(years,weights,c='r')
#Same plot with different color and line type and width
plt.plot(years,weights, 'r--',lw=3)
```

```python
# Bar plot for categorical x axis
x = ['Python', 'Java', 'C++', 'Javascript']
y = [45,20,30,5]
#plt.bar(x,y)
#Plot with different block color and edge color and alignment
plt.bar(x,y,color='r', align = 'edge', edgecolor = 'green')
```

```python
#Plot histogram
ages = np.random.normal(0,1,1000)
#plt.hist(ages)
#Increase the number of bins
#plt.hist(ages,bins=20)
#Plot cumulative distribution
plt.hist(ages,bins=20,cumulative=True)
```

```python
#Plot pie chart
plt.pie(y,labels=x,autopct="%.2f%%")
```

```python
# Box plot to plot the quartiles in a data
height = np.random.normal(178,8,300)
plt.boxplot(height)
```

## Customizing Plots

```python
years = [2014,2015,2016,2017,2018,2019,2020,2021]
income_A = [55,56,62,61,72,72,73,75]
income_B = [50,56,60,51,80,75,82,79]
plt.plot(years,income_A,label ="Person A")
plt.plot(years,income_B,label = "Person B")
plt.title("Income through years", fontsize = 20)
plt.xlabel("Time(years)")
plt.ylabel("Income($)")
plt.legend()
```
